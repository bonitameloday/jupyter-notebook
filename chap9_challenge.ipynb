```python
#01
from math import *
from random import *

# 반지름 구하기
r = random()*10
r1 = round(r, 2)

# 원의 면적 구하기
def getarea(r):
    s = pi * r**2
    return s

# 결과 출력
print('원의 반지름: ', r1)
print('원주율 pi: {:.4f}'.format(pi))
print('반지름 {}인 원의 면적은 {:.2f}'.format(r1, getarea(r)))
```

```python
#02
from math import *
from statistics import *
if __name__ == '__main__':
    # 팩토리얼 구하기
    print('1! = ', factorial(1))
    print('6! = ', factorial(6))
    print('11! = ', factorial(11))
    print('16! = ', factorial(16))
    print()
    
    # 중앙값, 평균, 분산, 표준편차 구하기
    st = [80, 99, 77, 65, 92, 74, 82]
    print(st)
    print('중앙 값: {:.2f}'.format(median(st)))
    print('평균: {:.2f}'.format(mean(st)))
    print('분산: {:.2f}'.format(variance(st)))
    print('표준편차: {:.2f}'.format(stdev(st)))
```

```python
#03
import turtle as t
colors = ['yellow', 'red', 'blue']
# 윈도 크기와 바탕색 설정
t.setup(500,500)
t.bgcolor('black')

# 태극 문양 그리기
for i in range(200):
    t.pencolor(colors[i%len(colors)])
    t.width(i/200+1)
    t.forward(i*2)
    t.left(119)
```

```python
#05
import turtle as t
from random import *

# 터틀 윈도 크기
t.setup(800, 500)

tshs = ['arrow', 'circle', 'turtle', 'square', 'triangle', 'classic']
han = ['화살', '원', '거북이', '사각형', '삼각형', '전통']
cols = ['red', 'blue', 'green', 'purple', 'magenta', 'black', 'gray', 'yellow', 'cyan', 'orange', 'aqua']

# 원 그리기 함수
def draw_circle():
    t.penup()
    t.goto(x, y)
    t.pendown()
    
    # 터틀의 색과 모양 변경, 터틀 모양 이름 한글의 색 변경
    turtle_shape = tshs.index(choice(tshs))
    turtle_color = choice(cols)
    text_color = choice(cols)
    
    t.color(turtle_color)
    t.shape(tshs[turtle_shape])
    
    t.begin_fill()
    t.circle(randint(3, 50))
    t.end_fill()
    
    t.penup()
    
    # 모양 이름과 색상 출력을 위해 위치 이동
    t.forward(20)  
    t.pendown()
    
    # 텍스트 색상을 무작위로 설정
    t.color(text_color)  
    t.write(f"{han[turtle_shape]}", align="center", font=("Arial", 10, "normal"))
    
    # 원래 위치로 되돌아감
    t.backward(20)  

# 20개의 원 그리기
for _ in range(20):
    x = randint(-300, 300)
    y = randint(-200, 200)
    draw_circle()

# 종료 전 터틀 창 유지
t.done()
```

```python
#06
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi * 2, np.pi * 2, 720)
cos, sin = np.cos(x), np.sin(x)

plt.plot(x, sin, ls = ':', label = 'sin')
plt.plot(x, cos, ls = '--', label = 'cos')

plt.title("Graph of sin cos")
plt.xlabel('radians')
plt.ylabel('value')
plt.grid()
plt.legend()
plt.show()
```
